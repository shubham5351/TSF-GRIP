# Exploring Decision tree

Q)In this given Iris dataset task is to construct the DecisionTreeClassifier and visualize it graphically.In this we have trained the model and while giving input parameters it will be able to classify under which category this belong. 

```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns
import sklearn.datasets as datasets
```

```python
#import iris dataset from sklearn library
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
print(iris_df.head()) #Print first 5 records

y = iris.target
print(y)
```

```python
#Representation of Statistical Data
iris_df.describe()
```

We are now checking for the datatype of each columns in the DataFrame and checking null values in it.

```python
iris_df.info()
```

<h3> Data Visualizations </h3>

Visualizing the count for each species.

```python
sns.countplot(x = y, data=iris_df)
plt.title('No of Species',fontsize=20)
plt.grid()
plt.xlabel('Species->')
sns.despine(top = True, right = True)
```

<h3> Preparing data</h3>

```python
#divide the data into input and output features
X = iris_df.iloc[:,].values
```

<h3> Training the Model </h3>

```python
#Defining DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
Dtree=DecisionTreeClassifier()
Dtree.fit(iris_df,y) #Fitting the model
```

<h3> Making Predictions </h3>

```python
sl = float(input("Enter Sepal Length:"))
sw = float(input("Enter Sepal width:"))
pl = float(input("Enter Petal length:"))
pw = float(input("Enter Petal width:"))
y_pred = Dtree.predict([[sl, sw, pl, pw]])
if y_pred == 0:
    %matplotlib inline
    img=mpimg.imread('Setosa.jpg')
    plt.figure(figsize=(5,10))
    plt.imshow(img)
elif y_pred == 1:
    %matplotlib inline
    img=mpimg.imread('Virginica.jpg')
    plt.figure(figsize=(5,10))
    plt.imshow(img)
else:
    %matplotlib inline
    img=mpimg.imread('Versicolor.jpg')
    plt.figure(figsize=(5,10))
    plt.imshow(img)
```

<h3>Visualizing Decision Tree </h3>

```python
#Text Representation of the Decision Tree
txt = tree.export_text(Dtree)
print(txt)
```

```python
#Visualize the Decision Tree
cn = ['setosa', 'versicolor', 'virginica']
plt.figure(figsize=(13,26))
tree.plot_tree(Dtree, feature_names = X, class_names = cn, filled = True);
```
